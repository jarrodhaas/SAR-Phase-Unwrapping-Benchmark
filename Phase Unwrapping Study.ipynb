# Phase Unwrapping

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def wrap(phases):
    return (phases + np.pi) % (2 * np.pi) - np.pi

def unwrap(w):
    
    result = []
    
    for i in range(w.shape[0]-1):
        grad = w[i+1] - w[i]
        if grad > np.pi:
            grad -= np.pi*2
        elif grad < -np.pi:
            grad += np.pi*2
            
        result.append(grad)
        
    grad = np.array(result)
        
    return np.concatenate(([0],np.cumsum(grad)), axis=0)

def makepath(length, rate):
    
    path = []
    curr = 1
    
    for i in range(length):
        step = (np.random.randint(rate) - ((rate-1)/2))
        
        curr += step
        
        path.append(curr)
        
    return np.array(path) + np.min(path)*-1
```

```python
path = 4*np.sin(np.linspace(0,10,30))

plt.plot(path)
```

```python
w = wrap(path)
plt.plot(w)
```

```python
pu = unwrap(w)
plt.plot(pu)
```
